```python
#ustawianie wartosci w kolumnie za pomoca NumPy
cars.loc[:,'cena'] = np.array([0] * len(cars))
cars
```

```python
#usuwanie kolumn, ktora posiadaja brakujace dane (w naszym wypadku nic nie zostanie usuniete)
cars.dropna(how='any')
#cars.fillna(value=...) - w tym miejscu moglibysmy uzupelnic nasze brakujace dane
```

```python
#zwrocenie sredniej danych
cars.mean()
#mozemy tez sprecyzowac, z ktorej osi chcemy pobrac nasze statystyki
cars.mean(1)
```

```python
#zwrocenie lacznej sumy naszych danych - do kazdego n rekordu, dodawane sa wartosci z rekordu n-1
cars.apply(np.cumsum)
```

```python
#grupowanie wedlug podanej kolumny - mozemy grupowac po wiekszej ilosci kolumn, przesylajac je jako liste
cars.groupby('marka').sum()
```

```python
#sortowanie wedlug podanej kolumny
cars.sort_values(by="marka")
```

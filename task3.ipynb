```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
cars = pd.read_csv('samochody1tys.csv')
```

```python
#utworzenie wykresu liniowego - do kazdego elementu z listy x przypisywana jest wartosc z listy y
x = [3,21,30]
y = [8,14,204]

#funkcja liniowa
plt.subplot(2,2,1)

plt.plot(x,y,'*--',label='liniowa')
plt.title('Wykres funkcji liniowej')
plt.xlabel('zbior x')
plt.ylabel('zbior y')
plt.legend()
plt.grid(True)

#funkcja kwadratowa
plt.subplot(2,2,2)

a = []
b = []

for x in range(-5,6):
    a.append(x)
    y = x**2
    b.append(y)

plt.plot(a,b,'o--',label='kwadratowa')
plt.title('Wykres funkcji liniowej')
plt.xlabel('zbior x')
plt.ylabel('zbior y')
plt.legend()
plt.grid(True)

#funkcja 3. rzedu
plt.subplot(2,2,3)
a = []
b = []
for x in range(-3,4):
    a.append(x)
    y = x**3
    b.append(y)
plt.plot(a,b,'x--',label='3. rzedu')
plt.title('Wykres funkcji 3. rzed')
plt.xlabel('zbior x')
plt.ylabel('zbior y')
plt.legend()
plt.grid(True)

#funkcja sinus
plt.subplot(2,2,4)
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y,label='sinus')
plt.title('Wykres funkcji sinus')
plt.xlabel('zbior x')
plt.ylabel('zbior y')
plt.legend()
plt.grid(True)

plt.subplots_adjust(top=4, bottom=2, left=2, right=3)

plt.show()
```

```python
#wykres slupkowy na podstawie ilosci powtarzajacych sie marek
marki = cars['marka']
df = marki.value_counts()
ax = df.plot.bar(figsize=(10,5))
plt.show()
```

```python
#wykres kolowy na podstawie powtarzajacych sie lat produkcji
rok_produkcji = cars['rok_produkcji']
df = rok_produkcji.value_counts()
ax = df.plot.pie(figsize=(20,20))
```

```python
#histogram - pokazuje czestotliwocs wystapienia danych wartosci w kolekcji
modele = cars['model']
df = modele.value_counts()
hist_plot = df.hist(bins=20)
hist_plot.set_title('Histogram modeli')
hist_plot.set_xlabel('ilosc')
hist_plot.set_ylabel('Czestotliwosc')
```
